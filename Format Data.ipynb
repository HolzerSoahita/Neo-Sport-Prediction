```python
player_file_path='Data/nba_players_stats_salary_2017_2018.csv'
player_data=pd.read_csv(player_file_path)
player_data.columns
```

```python
player_data_modif=player_data.copy()
player_data_modif.head(5)
```

```python
# Get list of categorical variables
s = (player_data_modif.dtypes == 'object')
object_cols = list(s[s].index)

print("Categorical variables:")
print(object_cols)
```

```python
from sklearn.preprocessing import LabelEncoder
# Apply label encoder to each column with categorical data
label_encoder = LabelEncoder()
player_data_modif[object_cols[1]] = label_encoder.fit_transform(player_data_modif[object_cols[1]])
player_data_modif[object_cols[1]] = label_encoder.transform(player_data_modif[object_cols[1]])
player_data_modif.head(5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

y_modif = player_data_modif.SALARY
feature_names_modif = ['TEAM','AGE', 'GP', 'MIN', 'W', 'L', 'PTS', 'FGM','FGA','3PM','3PA','FTM','FTA','OREB','DREB','AST', 'TOV', 'STL', 'BLK']
X_modif = player_data_modif[feature_names_modif]
train_X_modif, val_X_modif, train_y_modif, val_y_modif = train_test_split(X_modif, y_modif, random_state=1)
my_model_modif = RandomForestClassifier(random_state=0).fit(train_X_modif, train_y_modif)
```

```python
import eli5
from eli5.sklearn import PermutationImportance

perm_modif = PermutationImportance(my_model_modif, random_state=1).fit(val_X_modif, val_y_modif)
eli5.show_weights(perm_modif, feature_names = val_X_modif.columns.tolist())
```
