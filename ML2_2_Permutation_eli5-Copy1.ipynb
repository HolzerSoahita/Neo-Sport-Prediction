```python
import import_ipynb# Need To import custom packages installation
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

player_data = pd.read_csv('Data/nba_players_stats_salary_2017_2018.csv')
y = player_data.SALARY
feature_names = ['AGE', 'GP', 'MIN', 'W', 'L', 'PTS', 'FGM','FGA','3PM','3PA','FTM','FTA','OREB','DREB','AST', 'TOV', 'STL', 'BLK']
X = player_data[feature_names]
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
my_model = RandomForestClassifier(random_state=0).fit(train_X, train_y)
```

```python
import eli5
from eli5.sklearn import PermutationImportance

perm = PermutationImportance(my_model, random_state=1).fit(val_X, val_y)
eli5.show_weights(perm, feature_names = val_X.columns.tolist())
```
