```python
from __future__ import print_function
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
player_file_path='Data/nba_players_stats_salary_2017_2018.csv'
player_data_original=pd.read_csv(player_file_path)
player_data_original.head(5)
```

```python
player_data=player_data_original.copy()
player_data.head(5)
```

```python
# Get list of categorical variables
s = (player_data.dtypes == 'object')
object_cols = list(s[s].index)

print("Categorical variables:")
print(object_cols)
```

```python
from sklearn.preprocessing import LabelEncoder
# Apply label encoder to each column with categorical data
label_encoder = LabelEncoder()
player_data[object_cols[1]] = label_encoder.fit_transform(player_data[object_cols[1]])
player_data[object_cols[1]] = label_encoder.transform(player_data[object_cols[1]])
player_data.head(5)
```

```python
# Choose target and features
y = player_data.SALARY
player_features = ['AGE','FGM','OREB','FTA','GP','FGA','3PM','MIN','W','BLK','STL','PTS','TOV','L','FTM','3PA', 'AST','TEAM','DREB' ]
X = player_data[player_features]
```

```python
#Cross validation
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer


my_pipeline = Pipeline(steps=[('preprocessor', SimpleImputer()),
                              ('model', RandomForestRegressor(n_estimators=75,
                                                              random_state=0))
                             ])
```

```python
from sklearn.model_selection import cross_val_score

# Multiply by -1 since sklearn calculates *negative* MAE
scores = -1 * cross_val_score(my_pipeline, X, y,
                              cv=10,
                              scoring='neg_mean_absolute_error')

print("MAE scores:\n", scores)
print("Average MAE score (across experiments):")
print(scores.mean())
```
